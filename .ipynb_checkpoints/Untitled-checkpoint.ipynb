```python
import pclpy
from pclpy import pcl
import numpy as np

# 读取第一个点云
cloud1 = pcl.PointCloud.PointXYZRGB()
reader1 = pcl.io.PLYReader()
reader1.read("E:/SignalProccessing/ExperimentCode/icpSparse_MEE_MCC/data_gb/bunny_side/bunny_side_gb_0.78/bunny_side2_gb_0.78.ply", cloud1)

# 将第一个点云设置为蓝色
blue_color = pcl.visualization.PointCloudColorHandlerCustom.PointXYZRGB(cloud1, 0, 0, 255)

# 读取第二个点云
cloud2 = pcl.PointCloud.PointXYZRGB()
reader2 = pcl.io.PLYReader()
reader2.read("E:/SignalProccessing/ExperimentCode/icpSparse_MEE_MCC/data_gb/bunny_side/bunny_side_gb_0.78/bunny_side2_rotatey_180_gb_0.78.ply", cloud2)

# 将第二个点云设置为红色
red_color = pcl.visualization.PointCloudColorHandlerCustom.PointXYZRGB(cloud2, 255, 0, 0)

# 创建可视化器
viewer = pcl.visualization.PCLVisualizer("PCL Viewer")

# 设置背景颜色，将背景调为白色
viewer.setBackgroundColor(1, 1, 1)

# 添加第一个点云
viewer.addPointCloud(cloud1, blue_color, "cloud1")

# 添加第二个点云
viewer.addPointCloud(cloud2, red_color, "cloud2")

# 设置点云渲染大小
viewer.setPointCloudRenderingProperties(0, 1, "cloud1")
viewer.setPointCloudRenderingProperties(0, 1, "cloud2")

# 构建第二个点云的KdTree
kdtree = pcl.search.KdTree.PointXYZRGB()
kdtree.setInputCloud(cloud2)

# 找到第一个点云中每个点的最近邻
index= list(range(3750))

# 从生成的索引中随机选择1500个
selected_index = list(np.random.choice(index, size=750, replace=False))
for i in selected_index:
    query_point = cloud1.points[i]  # 获取点的坐标

    # 设置最近邻数量（这里设置为1，即找到一个最近邻）
    k = 1
    k_indices = pcl.vectors.Int()
    k_sqr_distances = pcl.vectors.Float()
    kdtree.nearestKSearch(query_point, k, k_indices, k_sqr_distances)
    #indices, _ = kdtree.nearestKSearch(query_point, k)

    # 获取匹配点的索引
    match_index = k_indices[0]

    # 获取匹配点的坐标
    match_point = cloud2.points[match_index]

    # 添加黑色线连接匹配对
    line_id = f"line_{i}"  # 使用 i 创建唯一 id
    viewer.addLine(query_point, match_point, 0, 1, 0, line_id)

while not viewer.wasStopped():
    viewer.spinOnce(10)
```
